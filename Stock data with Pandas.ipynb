# Downloading and clustering stock data in Pandas

In this section I will show how to download stock data using Pandas and use in-built functions in Scipy to sort the stocks according to how similarly their prices evolve. The group of Mantegna and others were among the first to show that stocks in the same industry tend to move similarly to each other. We can therefore build a sorted correlation matrix that shows the structure of the market.

In particular, I will calculate the logarithmic returns of a basket of stocks and sort the basket according to how the log-returns evolve. At the end I will draw a dendrogram and sorted correlation matrix.

Above I have two lists of stocks in the Dow Jones 30 and the S&P 100. There are some problems with the stock downloader; depending on the date range, you may not be able to download time series data for some of the stocks.

## Merging columns in Pandas

Here I will make a function that downloads stock data for a single metric, for example, adjusted close price. A second function constructs a Pandas dataframe whose columns are the adjusted close price for all the stocks in a given basket.

```python
stockList = spList  

# Obtains stock data for the metric you want
def stockMetric(stock,metric, start, end):
    return web.get_data_yahoo(str(stock), start, end)[metric]

# Construct the matrix of values for the selected metric
def metricMatrix(stockList,metric,start,end):
    old = stockMetric(stockList[0],metric, start,end)
    for i in range(1,len(stockList)):
        new = stockMetric(stockList[i],metric, start,end)
        old = pd.concat([old, new], axis = 1)
    return old
```

Often, the logarithmic returns
$r_i(t)\equiv \log P_i(t) - \log P_i(t-\tau)$ seems to be useful for people who analyse this type of thing. This can be done quite conveniently in Pandas:

```python
start = datetime.datetime(2011, 1, 1)
end = datetime.datetime(2013, 1, 1)

# Logarithmic returns matrix
def returns(stock,metric):
    stock[str(metric)]=np.log(stock[str(metric)])-np.log(stock[str(metric)].shift(1))
    return stock

myData = metricMatrix(spList[:50], 'Adj Close', start, end)

# Generate the logarithmic return for each stock that we have
returnsData = returns(myData,'Adj Close')
```

## Correlation matrix

There are several ways to create a correlation matrix in Python. Below I have calculated the Euclidean distance metric and displayed it as a square matrix. Note that the pdist function allows us to choose an in-built distance metric, or even better, define our own. From my recollection Matlab is a lot less flexible on the choice of distance metric.

```python
from scipy.spatial.distance import pdist, squareform

distance = squareform(pdist(returnsData.dropna(how='all').transpose(), metric='euclidean'))
%matplotlib inline
plt.matshow(distance, aspect='auto', origin='lower')
```

## Cluster analysis on the distance matrix

As above, the pdist function calculates the distance based on a metric you have selected or defined. Unfortunately, it appears that the linkage analysis in Scipy does not support square matrices. 

Also, note that the 'dropna' argument is particularly important to remove any numerical artifacts. There are other ways of dealing with the numerical artifacts. For instance, 'fillna' will substitute the NaNs in your distance matrix with a number of your choice (say 0 or 2, which is the maximum possible distance). 

```python
import scipy.spatial.distance as ssd
distanceCondensed = ssd.squareform(pdist(returnsData.dropna(how='all').transpose(), metric='correlation'))
%matplotlib inline
plt.matshow(distanceCondensed, aspect='auto', origin='lower')
plt.title('Distance metric in no particular order')
```

The linkage and dendrogram functions carry out the clustering analysis. In particular, the dendrogram function sorts the stocks as a function of their distance metric. We then arrange the 'distanceCondensed' matrix according to the ordering obtained in the clustering analysis (leaves in the dendrogram).

Here, I use the linkage function to perform a complete linkage clustering analysis on the distance matrix (calculated from the logarithmic returns). The dendrogram function provides the order in which we should shuffle the matrix. 

```python
from scipy.cluster.hierarchy import linkage, dendrogram

reorderedData = linkage(distanceCondensed, method='complete')
dendroData = dendrogram(reorderedData, orientation='right')

index = dendroData['leaves']
distanceCondensed = distanceCondensed[index,:]
distanceCondensed = distanceCondensed[:,index]
plt.matshow(distanceCondensed, aspect='auto', origin='lower', cmap='jet')
plt.title('Distance metric by complete linkage')
```
